# Global Tonemapping Operators

```python
% matplotlib inline

from __future__ import division

import matplotlib.pyplot
import numpy as np
import pylab
from OpenImageIO import FLOAT, ImageInput

import colour
from colour.plotting import *

EV = 2 ** 10
SAMPLES = np.linspace(0, EV, EV * 100)


def read_image_as_array(path, bit_depth=FLOAT):
    image = ImageInput.open(path)
    specification = image.spec()

    shape = (specification.height,
             specification.width,
             specification.nchannels)

    return np.array(image.read_image(bit_depth)).reshape(shape)


HDRI_IMAGE = read_image_as_array(
    '/colour-science/colour-ramblings/resources/images/ennis_small.exr')


def _srgb_transfer_function_vectorise(value):
    value = np.asarray(value)

    return np.where(value <= 0.0031308,
                    value * 12.92,
                    1.055 * (value ** (1 / 2.4)) - 0.055)


def image_plot(image,
               transfer_function=_srgb_transfer_function_vectorise):
    image = np.clip(transfer_function(image), 0, 1)
    pylab.imshow(image)

    settings = {'no_ticks': True,
                'bounding_box': [0, 1, 0, 1],
                'bbox_inches': 'tight',
                'pad_inches': 0}

    canvas(**{'figure_size': (16, 16)})
    decorate(**settings)
    display(**settings)
    
def tonemapping_operator_curves_plot(samples, curves, **kwargs):
    cycle = colour_cycle('rainbow', count=len(curves))
    for name, curve in sorted(curves.items()):
        pylab.plot(samples, curve, color=next(cycle), label=name)

    settings = {
        'x_label': '$log_2$ Input Luminance',
        'y_label': 'Linear Output Luminance',
        'legend': True,
        'legend_location': 'lower right',
        'x_ticker': True,
        'y_ticker': True,
        'x_ticker_locator': matplotlib.ticker.AutoMinorLocator(0.5),
        'grid': True,
        'x_tighten': True,
        'y_tighten': True,
        'limits': [0, np.max(samples), 0, 1]}
    settings.update(kwargs)

    matplotlib.pyplot.gca().set_xscale('log', basex=2)
    matplotlib.pyplot.gca().xaxis.set_major_formatter(
        matplotlib.ticker.ScalarFormatter())

    canvas(**{'figure_size': (8, 8)})
    decorate(**settings)
    boundaries(**settings)
    display(**settings)
```

## Simple Tonemapping Operator

```python
def simple_tonemapping_operator(RGB):
    return RGB / (RGB + 1)

tonemapping_operator_curves_plot(
    SAMPLES,
    {'Default': simple_tonemapping_operator(SAMPLES)},
    title='Simple Tonemapping Operator')
```

```python
image_plot(HDRI_IMAGE)
image_plot(simple_tonemapping_operator(HDRI_IMAGE))
```

## Filming Tonemapping Operator

```python
def filming_tonemapping_function(x, A, B, C, D, E, F):
    return ((x * (A * x + C * B) + D * E) / (x * (A * x + B) + D * F)) - E / F


def filming_tonemapping_operator(RGB,
                                 shoulder_strength=0.22,
                                 linear_strength=0.3,
                                 linear_angle=0.1,
                                 toe_strength=0.2,
                                 toe_numerator=0.01,
                                 toe_denominator=0.3,
                                 exposure_bias=2,
                                 linear_whitepoint=11.2):
    A = shoulder_strength
    B = linear_strength
    C = linear_angle
    D = toe_strength
    E = toe_numerator
    F = toe_denominator

    RGB = filming_tonemapping_function(RGB * exposure_bias, A, B, C, D, E, F)
    RGB *= 1 / filming_tonemapping_function(
        linear_whitepoint, A, B, C, D, E, F)
    RGB = np.clip(RGB, 0, 1)

    return RGB


tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Shoulder Strength - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, shoulder_strength=x))
         for x in np.linspace(0.1, 0.9, 9)),
    title='Filming Tonemapping Operator - Shoulder Strength')

tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Linear Strength - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, linear_strength=x))
         for x in np.linspace(0.1, 0.9, 9)),
    title='Filming Tonemapping Operator - Linear Strength')

tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Linear Angle - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, linear_angle=x))
         for x in np.linspace(0.1, 0.9, 9)),
    title='Filming Tonemapping Operator - Linear Angle')

tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Toe Strength - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, toe_strength=x))
         for x in np.linspace(0.1, 0.9, 9)),
    title='Filming Tonemapping Operator - Toe Strength')

tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Toe Numerator - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, toe_numerator=x))
         for x in np.linspace(0.01, 0.09, 9)),
    title='Filming Tonemapping Operator - Toe Numerator')

tonemapping_operator_curves_plot(
    SAMPLES,
    dict(('Toe Denominator - {0}'.format(x),
      filming_tonemapping_operator(SAMPLES, toe_denominator=x))
         for x in np.linspace(0.1, 0.9, 9)),
    title='Filming Tonemapping Operator - Toe Denominator')
```

```python
image_plot(HDRI_IMAGE)
image_plot(filming_tonemapping_operator(HDRI_IMAGE,
                                        shoulder_strength=0.05,
                                        linear_strength=0.75,
                                        linear_angle=0.45,
                                        toe_strength=0.2,
                                        toe_numerator=0.01,
                                        toe_denominator=0.3,
                                        exposure_bias=0.1))
```
