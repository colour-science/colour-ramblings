# RGB Colourspaces Transformations matrices

```python
import numpy as np
from IPython.html import widgets
from IPython.display import HTML, display
from collections import OrderedDict

import colour
from colour.utilities.verbose import message_box


# Rec. 2020 +
REC_2020_PLUS_PRIMARIES = np.array([[0.728, 0.294],
                                    [0.16490, 0.830],
                                    [0.1290, 0.0430]])
REC_2020_PLUS_WHITEPOINT = (0.32168, 0.33767)
REC_2020_PLUS_ILLUMINANT = 'D60'
REC_2020_PLUS_RGB_TO_XYZ_MATRIX = colour.normalised_primary_matrix(
    REC_2020_PLUS_PRIMARIES, REC_2020_PLUS_WHITEPOINT)
REC_2020_PLUS_XYZ_TO_RGB_MATRIX = np.linalg.inv(
    REC_2020_PLUS_RGB_TO_XYZ_MATRIX)

REC_2020_PLUS_COLOURSPACE = (
    colour.RGB_Colourspace('Rec. 2020 +',
                           REC_2020_PLUS_PRIMARIES,
                           REC_2020_PLUS_WHITEPOINT,
                           REC_2020_PLUS_ILLUMINANT,
                           REC_2020_PLUS_RGB_TO_XYZ_MATRIX,
                           REC_2020_PLUS_XYZ_TO_RGB_MATRIX))

colour.RGB_COLOURSPACES['Rec. 2020 +'] = REC_2020_PLUS_COLOURSPACE


def RGB_to_RGB(c_i, c_o, transform='Bradford'):
    cat = colour.chromatic_adaptation_matrix_VonKries(
        colour.xy_to_XYZ(c_o.whitepoint),
        colour.xy_to_XYZ(c_i.whitepoint),
        transform)

    return (np.dot(c_i.XYZ_to_RGB_matrix,
                   np.dot(cat, c_o.RGB_to_XYZ_matrix)))


def set_ui(colourspaces, cat):
    container = widgets.ContainerWidget()
    display(container)

    container.remove_class('vbox')
    container.add_class('hbox')

    c_i_DropdownWidget = widgets.DropdownWidget()
    c_i_DropdownWidget.values = colourspaces

    c_o_DropdownWidget = widgets.DropdownWidget()
    c_o_DropdownWidget.values = colourspaces

    cat_DropdownWidget = widgets.DropdownWidget()
    cat_DropdownWidget.values = cat

    return (container,
            c_i_DropdownWidget,
            c_o_DropdownWidget,
            cat_DropdownWidget)


container, c_i_DropdownWidget, c_o_DropdownWidget, cat_DropdownWidget = set_ui(
    OrderedDict([(v.name, v)
                 for k, v in
                 sorted(colour.RGB_COLOURSPACES.items())]),
    OrderedDict([(k, k)
                 for k, v in
                 sorted(colour.CHROMATIC_ADAPTATION_TRANSFORMS.items())]))


def compute_RGB_transformation_matrix(button):
    c_i, c_o = c_i_DropdownWidget.value, c_o_DropdownWidget.value
    cat = cat_DropdownWidget.value
    print cat
    message_box('{0} to {1}'.format(c_i.name, c_o.name))
    print(RGB_to_RGB(c_i_DropdownWidget.value, c_o_DropdownWidget.value, cat))
    print('\n')


compute_RGB_transformation_matrix_ButtonWidget = widgets.ButtonWidget(
    description="Compute RGB Transformation Matrix!")
compute_RGB_transformation_matrix_ButtonWidget.set_css({'margin': '5px'})
compute_RGB_transformation_matrix_ButtonWidget.on_click(
    compute_RGB_transformation_matrix)

container.children = [c_i_DropdownWidget,
                      c_o_DropdownWidget,
                      cat_DropdownWidget,
                      compute_RGB_transformation_matrix_ButtonWidget]
```

```python
def transformation_matrices(colourspaces, transform='Bradford'):
    matrices = {}
    for c_i in colourspaces:
        matrices[c_i.name] = {}
        for c_o in colourspaces:
            matrices[c_i.name][c_o.name] = RGB_to_RGB(c_i, c_o, transform)
    return matrices


def HTML_format_matrix(M, precision=7):
    # Handling whitepoint tuples.
    if type(M) is tuple:
        M = np.array(M).reshape(1, 2)

    html = '<table class="matrix-table">'
    shape = M.shape
    for i in range(shape[0]):
        html += '<tr class="matrix-row">'
        for j in range(shape[1]):
            v = M[i][j]
            pretty = '{{: 0.{}f}}'.format(precision).format(
                v) if precision is not None else v
            html += '<td class="matrix-column">{0}</td>'.format(pretty)
        html += '</tr>'
    html += '</table>'
    return html


def html_format_transformation_matrices(matrices, precision=7):
    colourspaces = sorted(matrices.values()[0].keys())
    
    html = ('<table class="table">'
            '<thead class="table-header">'
            '<tr class="table-row-header-group">'
            '<th class="table-canton"></th>')
    for colourspace in colourspaces:
        html += '<th class="table-row-header">{0}</th>'.format(colourspace)
    html += ('</tr>'
             '</thead>'
             '<tbody class="table-body">')
    for i, c_i in enumerate(colourspaces):
        html += '<tr class="table-row {0}">'.format('table-row-odd' if i % 2 == 0 else '')
        html += '<td class="table-data-header">{0}</td>'.format(c_i)
        for c_o in colourspaces:
            html += '<td class="table-data">{0}</td>'.format(HTML_format_matrix(matrices[c_i][c_o], precision))
        html += '</tr>'
    html += ('</tbody>'
             '</table>')
    return html


HTML(html_format_transformation_matrices(
    transformation_matrices([v for k, v in colour.RGB_COLOURSPACES.items()])))
```
