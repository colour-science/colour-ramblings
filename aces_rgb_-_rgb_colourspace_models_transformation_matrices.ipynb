# ACES RGB - RGB Colourspace Models Transformation Matrices - v0.0.1

The dataset is available here: https://github.com/colour-science/colour/tree/develop/colour/models/dataset

```python
import numpy as np
from IPython.display import HTML
from collections import namedtuple

import colour

# Rec. 2020 +
REC_2020_PLUS_PRIMARIES = np.array([[0.728, 0.294],
                                    [0.16490, 0.830],
                                    [0.1290, 0.0430]])
REC_2020_PLUS_WHITEPOINT = (0.32168, 0.33767)
REC_2020_PLUS_ILLUMINANT = 'D60'
REC_2020_PLUS_RGB_TO_XYZ_MATRIX = colour.normalised_primary_matrix(
    REC_2020_PLUS_PRIMARIES, REC_2020_PLUS_WHITEPOINT)
REC_2020_PLUS_XYZ_TO_RGB_MATRIX = np.linalg.inv(
    REC_2020_PLUS_RGB_TO_XYZ_MATRIX)

REC_2020_PLUS_COLOURSPACE = (
    colour.RGB_Colourspace('Rec. 2020 +',
                           REC_2020_PLUS_PRIMARIES,
                           REC_2020_PLUS_WHITEPOINT,
                           REC_2020_PLUS_ILLUMINANT,
                           REC_2020_PLUS_RGB_TO_XYZ_MATRIX,
                           REC_2020_PLUS_XYZ_TO_RGB_MATRIX))

colour.RGB_COLOURSPACES['Rec. 2020 +'] = REC_2020_PLUS_COLOURSPACE

ACES_RGB = colour.RGB_COLOURSPACES['ACES RGB']

RGB_TransformationMatrices = namedtuple('RGB_TransformationMatrices',
                                        ('ACES_RGB_to_RGB_matrix',
                                         'RGB_to_ACES_RGB_matrix'))


def ACES_RGB_to_RGB(colourspace, transform='Bradford'):
    cat = colour.chromatic_adaptation_matrix_VonKries(
        colour.xy_to_XYZ(ACES_RGB.whitepoint),
        colour.xy_to_XYZ(colourspace.whitepoint),
        transform)

    return (np.dot(colourspace.XYZ_to_RGB_matrix,
                   np.dot(cat, ACES_RGB.RGB_to_XYZ_matrix)))


def RGB_to_ACES_RGB(colourspace, transform='Bradford'):
    return np.linalg.inv(ACES_RGB_to_RGB(colourspace, transform))


def transformation_matrices(colourspaces, transform='Bradford'):
    matrices = {}
    for colourspace in colourspaces:
        matrices[colourspace.name] = RGB_TransformationMatrices(
            ACES_RGB_to_RGB(colourspace, transform),
            RGB_to_ACES_RGB(colourspace, transform))
    return matrices


def html_format_transformation_matrices(matrices, precision=7):
    def format_matrix(M, precision=precision):
        pretty = (lambda
                      x: '<span style="padding: 0 8px"></span>'.join(
            map('{{: 0.{}f}}'.format(precision).format, x)))

        return ('<td>{0}<br>{1}<br>{2}</td>').format(pretty(M[0]),
                                                     pretty(M[1]),
                                                     pretty(M[2]))

    html = ('<table><tr>'
            '<th style="text-align: center">Colourspace</th>'
            '<th style="text-align: center">'
            'ACES RGB to RGB</th>'
            '<th style="text-align: center">'
            'RGB to ACES RGB</th></tr>')
    for colourspace, (ACES_RGB_to_RGB_matrix,
                      RGB_to_ACES_RGB_matrix) in matrices.items():
        html += '<tr>'
        html += '<td style="text-align: center">{0}</td>'.format(colourspace)
        html += format_matrix(ACES_RGB_to_RGB_matrix)
        html += format_matrix(RGB_to_ACES_RGB_matrix)
        html += '</tr>'
    html += '</table>'
    return html
```

## VFX Colourspaces - Bradford CAT

```python
VFX_COLOURSPACES = ('ACES RGB',
                    'ALEXA Wide Gamut RGB',
                    'DCI-P3',
                    'Rec. 2020',
                    'Rec. 2020 +',
                    'Rec. 709',
                    'S-Log')

HTML(html_format_transformation_matrices(
    transformation_matrices(
        (colour.RGB_COLOURSPACES[c] for c in VFX_COLOURSPACES))))
```
