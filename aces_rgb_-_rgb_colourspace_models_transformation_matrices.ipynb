# ACES RGB - RGB Colourspace Models Transformation Matrices - v0.0.5

The dataset is available here: https://github.com/colour-science/colour/tree/feature/aces_wg/colour/models/dataset

```python
import numpy as np
from IPython.display import HTML
from collections import namedtuple

import colour

# Rec. 2020 +
REC_2020_PLUS_PRIMARIES = np.array([[0.728, 0.294],
                                    [0.16490, 0.830],
                                    [0.1290, 0.0430]])
REC_2020_PLUS_WHITEPOINT = (0.32168, 0.33767)
REC_2020_PLUS_ILLUMINANT = 'D60'
REC_2020_PLUS_RGB_TO_XYZ_MATRIX = colour.normalised_primary_matrix(
    REC_2020_PLUS_PRIMARIES, REC_2020_PLUS_WHITEPOINT)
REC_2020_PLUS_XYZ_TO_RGB_MATRIX = np.linalg.inv(
    REC_2020_PLUS_RGB_TO_XYZ_MATRIX)

REC_2020_PLUS_COLOURSPACE = (
    colour.RGB_Colourspace('Rec. 2020 +',
                           REC_2020_PLUS_PRIMARIES,
                           REC_2020_PLUS_WHITEPOINT,
                           REC_2020_PLUS_ILLUMINANT,
                           REC_2020_PLUS_RGB_TO_XYZ_MATRIX,
                           REC_2020_PLUS_XYZ_TO_RGB_MATRIX))

colour.RGB_COLOURSPACES['Rec. 2020 +'] = REC_2020_PLUS_COLOURSPACE

ACES_RGB = colour.RGB_COLOURSPACES['ACES RGB']

RGB_TransformationMatrices = namedtuple('RGB_TransformationMatrices',
                                        ('ACES_RGB_to_RGB_matrix',
                                         'RGB_to_ACES_RGB_matrix'))


def ACES_RGB_to_RGB(colourspace, transform='Bradford'):
    cat = colour.chromatic_adaptation_matrix_VonKries(
        colour.xy_to_XYZ(ACES_RGB.whitepoint),
        colour.xy_to_XYZ(colourspace.whitepoint),
        transform)

    return (np.dot(colourspace.XYZ_to_RGB_matrix,
                   np.dot(cat, ACES_RGB.RGB_to_XYZ_matrix)))


def RGB_to_ACES_RGB(colourspace, transform='Bradford'):
    return np.linalg.inv(ACES_RGB_to_RGB(colourspace, transform))


def transformation_matrices(colourspaces, transform='Bradford'):
    matrices = {}
    for colourspace in colourspaces:
        matrices[colourspace.name] = RGB_TransformationMatrices(
            ACES_RGB_to_RGB(colourspace, transform),
            RGB_to_ACES_RGB(colourspace, transform))
    return matrices


def HTML_format_matrix(M, precision=7):
    # Handling whitepoint tuples.
    if type(M) is tuple:
        M = np.array(M).reshape(1, 2)

    html = '<table class="matrix-table">'
    shape = M.shape
    for i in range(shape[0]):
        html += '<tr class="matrix-row">'
        for j in range(shape[1]):
            v = M[i][j]
            pretty = '{{: 0.{}f}}'.format(precision).format(
                v) if precision is not None else v
            html += '<td class="matrix-column">{0}</td>'.format(pretty)
        html += '</tr>'
    html += '</table>'
    return html


def html_format_transformation_matrices(matrices, precision=7):
    html = ('<table class="table">'
            '<thead class="table-header">'
            '<tr class="table-row-header-group">'
            '<th class="table-canton"></th>'
            '<th class="table-row-header">ACES RGB to RGB</th>'
            '<th class="table-row-header">RGB to ACES RGB</th>'
            '</tr>'
            '</thead>'
            '<tbody class="table-body">')
    for i, matrix in enumerate(sorted(matrices.items())):
        colourspace, (ACES_RGB_to_RGB_matrix, RGB_to_ACES_RGB_matrix) = matrix
        html += '<tr class="table-row {0}">'.format('table-row-odd' if i % 2 == 0 else '')
        html += '<td class="table-row-header">{0}</td>'.format(colourspace)
        html += '<td class="table-data">{0}</td>'.format(
            HTML_format_matrix(ACES_RGB_to_RGB_matrix, precision))
        html += '<td class="table-data">{0}</td>'.format(
            HTML_format_matrix(RGB_to_ACES_RGB_matrix, precision))
        html += '</tr>'
    html += ('</tbody>'
             '</table>')
    return html
```

## Nuke - Scripts Make

```python
NUKE_TEMPLATE = '''
BackdropNode {{
 inputs 0
 name {colourspace}
 tile_color 0x3f6c7fff
 note_font_size 42
 selected true
 xpos {backdrop_x}
 ypos -56
 bdwidth 170
 bdheight 140
}}

ColorMatrix {{
 matrix {{
     {ACES_RGB_to_RGB_matrix}
   }}
 name {ACES_RGB_to_RGB}
 label {label}
 selected true
 xpos {color_matrix_x}
 ypos 0
}}

ColorMatrix {{
 inputs 0
 matrix {{
     {RGB_to_ACES_RGB_matrix}
   }}
 name {RGB_to_ACES_RGB}
 label {label}
 selected true
 xpos {color_matrix_x}
 ypos 40
}}
'''


def make_nuke_script(matrices, label, precision=7):
    def nk_format_matrix(M, precision=precision):
        pretty = lambda x: ' '.join(
            map('{{:0.{}f}}'.format(precision).format, x))
        
        nk = '{{{0}}}\n'.format(pretty(M[0]))
        nk += '     {{{0}}}\n'.format(pretty(M[1]))
        nk += '     {{{0}}}'.format(pretty(M[2]))
        return nk

    x = 0
    script = ''
    for colourspace, (ACES_RGB_to_RGB_matrix,
                      RGB_to_ACES_RGB_matrix) in sorted(matrices.items()):
        name = colourspace.replace(' ', '_')
        script += NUKE_TEMPLATE.format(
            colourspace=name,
            ACES_RGB_to_RGB_matrix=nk_format_matrix(ACES_RGB_to_RGB_matrix),
            ACES_RGB_to_RGB='ACES_RGB_to_{0}'.format(name),
            RGB_to_ACES_RGB_matrix=nk_format_matrix(RGB_to_ACES_RGB_matrix),
            RGB_to_ACES_RGB='{0}_to_ACES_RGB'.format(name),
            label='"{0}"'.format(label),
            backdrop_x=x - 50,
            backdrop_width=x + 170,
            color_matrix_x=x)
        x += 200

    return script


with open('rgb_colourspaces_bradford.nk', 'w') as file:
    file.write(make_nuke_script(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items()))),
                                 'Bradford CAT'))

with open('rgb_colourspaces_cat02.nk', 'w') as file:
    file.write(make_nuke_script(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items())),
        'CAT02'),
                                 'CAT02 CAT'))

with open('rgb_colourspaces_bianco.nk', 'w') as file:
    file.write(make_nuke_script(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items())),
        'Bianco'),
                                'Bianco CAT'))
```

## LaTeX - Documents Make

```python
LATEX_TABLE_TEMPLATE = '''
\\documentclass{{article}}
\\usepackage{{adjustbox}}
\\let\\belowcaptionskip\\abovecaptionskip
\\begin{{document}}

\\begin{{table}}[h!]
\\caption{{{caption}}}
\\centering
\\begin{{adjustbox}}{{width=\\textwidth, totalheight=\\textheight, keepaspectratio}}
\\begin{{tabular}}{{|c|ccc|ccc|}}
\\hline
Colourspace & \\multicolumn{{3}}{{c|}}{{ACES RGB to RGB}} & \\multicolumn{{3}}{{c|}}{{RGB to ACES RGB}}\\\\
\\hline
\\hline
{transformation_matrices}
\\end{{tabular}}
\\end{{adjustbox}}
\\end{{table}}

\\end{{document}}
'''


def make_latex_table(matrices, caption, precision=7):
    pretty = lambda x: ' & '.join(
        map('{{:0.{}f}}'.format(precision).format, x))

    transformation_matrices = ''
    for colourspace, (ACES_RGB_to_RGB_matrix,
                      RGB_to_ACES_RGB_matrix) in sorted(matrices.items()):
        for i in range(3):
            transformation_matrices += colourspace if i == 1 else '{}'
            transformation_matrices += ' & '
            transformation_matrices += pretty(ACES_RGB_to_RGB_matrix[i])
            transformation_matrices += ' & '
            transformation_matrices += pretty(RGB_to_ACES_RGB_matrix[i])
            transformation_matrices += ' \\\\\n'
        transformation_matrices += '\\hline\n'

    table = LATEX_TABLE_TEMPLATE.format(
        caption=caption,
        transformation_matrices=transformation_matrices)
    return table


with open('rgb_colourspaces_bradford.tex', 'w') as file:
    file.write(make_latex_table(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items()))),
                                 'Bradford CAT'))

with open('rgb_colourspaces_cat02.tex', 'w') as file:
    file.write(make_latex_table(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items())),
        'CAT02'),
                                 'CAT02 CAT'))

with open('rgb_colourspaces_bianco.tex', 'w') as file:
    file.write(make_latex_table(transformation_matrices(
        (v for k, v in sorted(colour.RGB_COLOURSPACES.items())),
        'Bianco'),
                                'Bianco CAT'))
```

## RGB Colourspace Models Dataset

> Note: The dataset values are presented here as they have been retrieved in the various references.

```python
def html_format_colourspaces(colourspaces):
    html = ('<table class="table">'
            '<thead class="table-header">'
            '<tr class="table-row-header-group">'
            '<th class="table-canton"></th>'
            '<th class="table-row-header">Primaries</th>'
            '<th class="table-row-header">Whitepoint</th>'
            '<th class="table-row-header">Illuminant</th>'
            '<th class="table-row-header">XYZ to RGB</th>'
            '<th class="table-row-header">RGB to XYZ</th>'
            '</tr>'
            '</thead>'
            '<tbody>')
    for i, colourspace in enumerate(colourspaces):
        html += '<tr class="table-row {0}">'.format('table-row-odd' if i % 2 == 0 else '')
        html += '<td class="table-data-header">{0}</td>'.format(
            colourspace.name)
        html += '<td class="table-data">{0}</td>'.format(
            HTML_format_matrix(colourspace.primaries, precision=None))
        html += '<td class="table-data text-align-center vertical-align-top">{0}</td>'.format(
            HTML_format_matrix(colourspace.whitepoint, precision=None))
        html += '<td class="table-data text-align-center vertical-align-top">{0}</td>'.format(
            colourspace.illuminant)
        html += '<td class="table-data">{0}</td>'.format(
            HTML_format_matrix(colourspace.XYZ_to_RGB_matrix, precision=None))
        html += '<td class="table-data">{0}</td>'.format(
            HTML_format_matrix(colourspace.RGB_to_XYZ_matrix, precision=None))
        html += '</tr>'
    html += ('<tbody>'
             '</table>')
    return html


HTML(html_format_colourspaces(
    (v for k, v in sorted(colour.RGB_COLOURSPACES.items()))))
```
