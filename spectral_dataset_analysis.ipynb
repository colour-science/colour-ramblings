# Spectral Dataset Analysis

```python
import numpy as np
import os
import pandas
import scipy.io
from collections import namedtuple
from pprint import pprint

import colour

InterpolationData = namedtuple(
    'InterpolationData',
    ('name', 'shape', 'zeros',
     'linear', 'sprague', 'cubic', 'pchip',
     'd_E_linear', 'd_E_sprague', 'd_E_cubic', 'd_E_pchip'))

MatlabSpectralDataDescriptor = namedtuple(
    'MatlabSpectralDataDescriptor',
    ('name', 'key', 'shape', 'transpose', 'identifiers'))


def read_spds_from_mat_file(mat_file, descriptors, namespace_separator='|'):
    matlab_data = scipy.io.loadmat(mat_file)
    spds = {}
    for descriptor in descriptors:
        descriptor_functions = matlab_data[descriptor.key]
        wavelengths = descriptor.shape.range()

        if descriptor.transpose:
            descriptor_functions = np.transpose(descriptor_functions)

        for i, descriptor_function in enumerate(descriptor_functions):
            identifier = (i + 1 if descriptor.identifiers is None else
                          matlab_data[descriptor.identifiers][i].strip())
            name = '{0}{1}{2}'.format(descriptor.name, 
                                      namespace_separator, 
                                      identifier)
            spds[name] = colour.SpectralPowerDistribution(
                name,
                dict(zip(wavelengths, descriptor_function)))

    return spds

def add_spds_to_spectral_dataset(spds, dataset, prefix):
    for name, spd in spds.items():
        spd.name = '{0}|{1}'.format(prefix, name)
        dataset[spd.name] = spd

def spectral_dataset(dataset='All'):
    spectral_dataset_spds = {}

    # *Colour - Illuminants*
    add_spds_to_spectral_dataset(colour.ILLUMINANTS_RELATIVE_SPDS,
                                 spectral_dataset_spds,
                                 'Colour|Illuminants')

    # *Colour - Light Sources*
    add_spds_to_spectral_dataset(colour.LIGHT_SOURCES_RELATIVE_SPDS,
                                 spectral_dataset_spds,
                                 'Colour|Light Sources')

    # *Colour - Characterisation*
    for cc in ('BabelColor Average', 'ColorChecker N Ohta'):
        add_spds_to_spectral_dataset(colour.COLOURCHECKERS_SPDS[cc],
                                     spectral_dataset_spds,
                                     'Colour|Characterisation|{0}'.format(cc))

    # *Colour - Quality*
    add_spds_to_spectral_dataset(colour.TCS_SPDS,
                                 spectral_dataset_spds,
                                 'Colour|Quality|TCS')
    add_spds_to_spectral_dataset(colour.VS_SPDS,
                                 spectral_dataset_spds,
                                 'Colour|Quality|VS')

    # *Colour - Recovery*
    add_spds_to_spectral_dataset(colour.SMITS_1999_SPDS,
                                 spectral_dataset_spds,
                                 'Colour|Recovery|Smits (1999)')

    # *X-Rite*
    spds = colour.read_spds_from_xrite_file(
        'resources/others/Digital_ColorChecker_SG.txt')
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'X-Rite|Digital ColorChecker SG')

    spds = colour.read_spds_from_xrite_file(
        'resources/others/ColorChecker_DC.txt')
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'X-Rite|ColorChecker DC')

    # *Spectral Color Research Group*
    # *Agfa ColorReference IT8.7*    
    spds = read_spds_from_mat_file(
        'resources/others/matlab/agfait872.mat',
        [MatlabSpectralDataDescriptor(
            'Agfa ColorReference IT8.7', 'agfa',
            colour.SpectralShape(400, 700, 10),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Artificial Lights*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/Art_lights.mat',
        [MatlabSpectralDataDescriptor(
            'Artificial Lights', 'lights',
            colour.SpectralShape(380, 780, 2),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Candy Dyes*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/candy_matlab.mat',
        [MatlabSpectralDataDescriptor(
            'Candy Dyes', 'candy',
            colour.SpectralShape(190, 820, 2),
            True, 'S')])

    del spds['Candy Dyes|blank']

    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Munsell - Glossy - 1600*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/munsell380_780_1_glossy.mat',
        [MatlabSpectralDataDescriptor(
            'Munsell - Glossy - 1600', 'X',
            colour.SpectralShape(380, 780, 1),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Munsell - Matte - 1269*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/munsell380_800_1.mat',
        [MatlabSpectralDataDescriptor(
            'Munsell - Matte - 1269', 'munsell',
            colour.SpectralShape(380, 800, 1),
            True, 'S')])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Munsell - Matte - 1250*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/munsell400_700_5.mat',
        [MatlabSpectralDataDescriptor(
            'Munsell - Matte - 1250', 'munsell',
            colour.SpectralShape(400, 700, 5),
            True, 'S')])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Munsell - Glossy - 32*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/munsell400_700_10.mat',
        [MatlabSpectralDataDescriptor(
            'Munsell - Glossy - 32', 'munsell',
            colour.SpectralShape(400, 700, 10),
            True, 'S')])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Daylight - BaSO4*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/daylight/BaSO4.mat',
        [MatlabSpectralDataDescriptor(
            'Daylight - BaSO4', 'baso4',
            colour.SpectralShape(390, 1070, 4),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Daylight - Tree*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/daylight/tree.mat',
        [MatlabSpectralDataDescriptor(
            'Daylight - Tree', 'tree',
            colour.SpectralShape(390, 1070, 4),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    # *Daylight - Sky*
    spds = read_spds_from_mat_file(
        'resources/others/matlab/daylight/sky.mat',
        [MatlabSpectralDataDescriptor(
            'Daylight - Sky', 'sky',
            colour.SpectralShape(390, 1070, 4),
            True, None)])
    add_spds_to_spectral_dataset(spds,
                                 spectral_dataset_spds,
                                 'S.C.R.G.')

    return spectral_dataset_spds


def normalise_XYZ(XYZ):
    return XYZ / XYZ[1]


def interpolation_data(spds,
                       cmfs=colour.CMFS[
                           'CIE 1931 2 Degree Standard Observer'],
                       illuminant=colour.ILLUMINANTS[
                           'CIE 1931 2 Degree Standard Observer']['E'],
                       steps=1):
    interpolation_data = {}
    for name, spd in spds.items():
        # *Zeroes* computation / ground truth.
        try:
            XYZ_z = normalise_XYZ(colour.spectral_to_XYZ(
                spd, cmfs))
        except RuntimeError:
            XYZ_z = None

        shape = colour.SpectralShape(steps=steps)

        # *Linear* interpolation computation.
        XYZ_l = normalise_XYZ(colour.spectral_to_XYZ(
            spd.clone().interpolate(shape, method='Linear'),
            cmfs))
        # *Sprague* interpolation computation.
        XYZ_s = normalise_XYZ(colour.spectral_to_XYZ(
            spd.clone().interpolate(shape),
            cmfs))
        # *Cubic Spline* interpolation computation.
        XYZ_c = normalise_XYZ(colour.spectral_to_XYZ(
            spd.clone().interpolate(shape, method='Cubic Spline'),
            cmfs))
        # *Shape Preserving* interpolation computation.
        XYZ_p = normalise_XYZ(colour.spectral_to_XYZ(
            spd.clone().interpolate(shape, method='Shape Preserving'),
            cmfs))

        if XYZ_z is not None:
            Lab_z = colour.XYZ_to_Lab(XYZ_z, illuminant)
            d_E_XYZ_l = colour.delta_E_CIE2000(
                Lab_z,
                colour.XYZ_to_Lab(XYZ_l, illuminant))
            d_E_XYZ_s = colour.delta_E_CIE2000(
                Lab_z,
                colour.XYZ_to_Lab(XYZ_s, illuminant))
            d_E_XYZ_c = colour.delta_E_CIE2000(
                Lab_z,
                colour.XYZ_to_Lab(XYZ_c, illuminant))
            d_E_XYZ_p = colour.delta_E_CIE2000(
                Lab_z,
                colour.XYZ_to_Lab(XYZ_p, illuminant))
        else:
            d_E_XYZ_l = d_E_XYZ_s = d_E_XYZ_c = d_E_XYZ_p = None
        interpolation_data[name] = InterpolationData(
            name, (spd.shape.start, spd.shape.end, spd.shape.steps),
            XYZ_z, XYZ_l, XYZ_s, XYZ_c, XYZ_p,
            d_E_XYZ_l, d_E_XYZ_s, d_E_XYZ_c, d_E_XYZ_p)

    return interpolation_data
```
