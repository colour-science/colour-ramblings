# Spectral Dataset Analysis

```python
from pprint import pprint
from collections import namedtuple

import colour

XYZ_InterpolationData = namedtuple(
    'XYZ_InterpolationData',
    ('name', 'shape', 'zeros', 'sprague', 'cubic', 'linear'))


def colour_spectral_dataset(dataset='all'):
    spectral_dataset_spds = {}
    if dataset in ('all', 'illuminants'):
        spectral_dataset_spds.update(colour.ILLUMINANTS_RELATIVE_SPDS)

    if dataset in ('all', 'light_sources'):
        spectral_dataset_spds.update(colour.LIGHT_SOURCES_RELATIVE_SPDS)

    if dataset in ('all', 'cmfs'):
        for cmfs in colour.CMFS.values():
            if cmfs.name in ('cie_2_1931', 'cie_10_1964'):
                continue

            for axis in ('x', 'y', 'z'):
                spd = getattr(cmfs, axis).clone()
                spd.name = '{0}|{1}'.format(cmfs.name, spd.name)
                spectral_dataset_spds[spd.name] = spd

    if dataset in ('all', 'characterisation'):
        for cc in ('BabelColor Average', 'ColorChecker N Ohta'):
            for name, spd in colour.COLOURCHECKERS_SPDS[cc].items():
                spd = spd.clone()
                spd.name = '{0}|{1}'.format(cc, name)
                spectral_dataset_spds[spd.name] = spd

    if dataset in ('all', 'quality'):
        spectral_dataset_spds.update(colour.TCS_SPDS)
        spectral_dataset_spds.update(colour.VS_SPDS)

    if dataset in ('all', 'recovery'):
        spectral_dataset_spds.update(colour.SMITS_1999_SPDS)

    if dataset in ('all', 'tabular'):
        spds = colour.read_spds_from_xrite_file(
            'resources/others/Digital_ColorChecker_SG.txt')
        for name, spd in spds.items():
            spd.name = '{0}|{1}'.format('Digital ColorChecker SG', name)
            spectral_dataset_spds[spd.name] = spd
        spds = colour.read_spds_from_xrite_file(
            'resources/others/ColorChecker_DC.txt')
        for name, spd in spds.items():
            spd.name = '{0}|{1}'.format('ColorChecker DC', name)
            spectral_dataset_spds[spd.name] = spd

    return spectral_dataset_spds


def XYZ_interpolation_data(spds,
                           cmfs=colour.CMFS[
                               'CIE 1931 2 Degree Standard Observer']):
    interpolation_data = {}
    for name, spd in spds.items():
        try:
            XYZ_z = colour.spectral_to_XYZ(
                spd, cmfs)
        except RuntimeError:
            XYZ_z = None
            
        XYZ_s = colour.spectral_to_XYZ(
            spd.clone().interpolate(colour.SpectralShape(steps=1)),
            cmfs)
        XYZ_c = colour.spectral_to_XYZ(
            spd.clone().interpolate(colour.SpectralShape(steps=1),
                                    method='Cubic Spline'),
            cmfs)
        XYZ_l = colour.spectral_to_XYZ(
            spd.clone().interpolate(colour.SpectralShape(steps=1),
                                    method='Linear'),
            cmfs)
        interpolation_data[name] = XYZ_InterpolationData(
            name, spd.shape, XYZ_z, XYZ_s, XYZ_c, XYZ_l)

    return interpolation_data

pprint(XYZ_interpolation_data(colour_spectral_dataset()))
```
